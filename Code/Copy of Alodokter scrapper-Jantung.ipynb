```python
import requests
from bs4 import BeautifulSoup
from csv import reader, writer
from time import sleep
import lxml
import json
```

```python
topics = [
    'hipertensi'
]
```

```python
links = [] # tuple of (topic, url)

for topic in topics:
    #time.sleep(1)
    curr_links = []
    page = 11
    
    while True :
        search_page = requests.get('https://www.alodokter.com/komunitas/topic-tag/' + topic + '/page/{}'.format(page))
        if search_page.status_code == 404:
            break
        else:
            tries_text =  BeautifulSoup(search_page.text, 'lxml')
            card_topics = tries_text.find_all('card-topic')
            results = filter(lambda card: card['answered-doctor'] == 'true', card_topics)
            results = list(map(lambda card: card['href'], results))
            curr_links.extend(results)
            page += 1
    if curr_links:
        final_links = [(topic,  url) for url in curr_links]
        links.extend(final_links)    
```

```python
HOMEPAGE = 'https://www.alodokter.com'
```

```python
query = list(map(lambda link: requests.get(HOMEPAGE + link[1]).text, links))
```

```python
queryy = list(map(lambda resp: BeautifulSoup(resp, 'lxml'), query))
question = list(map(lambda resp: resp.find('detail-topic')['member-topic-content'], queryy[:1155] + queryy[1156:]))
#answer = list(map(lambda resp: resp.find('doctor-topic')['doctor-topic-content'], query))
question = list(map(lambda resp: BeautifulSoup(json.loads(resp), 'lxml').text, question))
#answer = list(map(lambda resp: BeautifulSoup(json.loads(resp), 'lxml').text, answer))
```
